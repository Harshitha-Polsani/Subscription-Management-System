```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import mysql.connector
from mysql.connector import Error

def customer():
    def insert_customer():
        customer_id = entry_customer_id.get()
        first_name = entry_first_name.get()
        last_name = entry_last_name.get()
        address = entry_address.get()
        
        if not first_name or not last_name or not address:
            messagebox.showerror("Error", "All fields must be filled")
            return
        
        try:
            connection = connect_to_database()
            if connection:
                cursor = connection.cursor()
                # Check if customer with the same ID already exists
                if customer_id:
                    cursor.execute("SELECT IdNo FROM CUSTOMER WHERE IdNo = %s", (customer_id,))
                    existing_customer = cursor.fetchone()
                    if existing_customer:
                        messagebox.showerror("Error", "Customer with the same ID already exists")
                        return
                # Insert customer record
                insert_query = "INSERT INTO CUSTOMER (IdNo, F_Name, L_Name, Address) VALUES (%s, %s, %s, %s)"
                cursor.execute(insert_query, (customer_id, first_name, last_name, address))
                connection.commit()
                messagebox.showinfo("Success", "Customer record inserted successfully")
                clear_entries()
        except Error as e:
            messagebox.showerror("Error", f"Error while inserting customer record: {e}")

    def update_customer():
        customer_id = entry_customer_id.get()
        first_name = entry_first_name.get()
        last_name = entry_last_name.get()
        address = entry_address.get()
        
        if not customer_id or not first_name or not last_name or not address:
            messagebox.showerror("Error", "All fields must be filled")
            return
        
        try:
            connection = connect_to_database()
            if connection:
                cursor = connection.cursor()
                # Check if customer with the given ID exists
                cursor.execute("SELECT IdNo FROM CUSTOMER WHERE IdNo = %s", (customer_id,))
                existing_customer = cursor.fetchone()
                if not existing_customer:
                    messagebox.showerror("Error", "Customer with the provided ID does not exist")
                    return
                # Update customer record
                update_query = "UPDATE CUSTOMER SET F_Name = %s, L_Name = %s, Address = %s WHERE IdNo = %s"
                cursor.execute(update_query, (first_name, last_name, address, customer_id))
                connection.commit()
                messagebox.showinfo("Success", "Customer record updated successfully")
                clear_entries()
        except Error as e:
            messagebox.showerror("Error", f"Error while updating customer record: {e}")

    def delete_customer():
        customer_id = entry_customer_id.get()
        first_name = entry_first_name.get()
        last_name = entry_last_name.get()
        address = entry_address.get()
        
        if not customer_id and not first_name and not last_name and not address:
            messagebox.showerror("Error", "Specify at least one detail to identify the customer")
            return
        
        try:
            connection = connect_to_database()
            if connection:
                cursor = connection.cursor()
                # Check if customer with the provided details exists
                if customer_id:
                    cursor.execute("SELECT IdNo FROM CUSTOMER WHERE IdNo = %s", (customer_id,))
                else:
                    cursor.execute("SELECT IdNo FROM CUSTOMER WHERE F_Name = %s AND L_Name = %s AND Address = %s",
                                   (first_name, last_name, address))
                existing_customer = cursor.fetchone()
                if not existing_customer:
                    messagebox.showerror("Error", "Customer with the provided details does not exist")
                    return
                # Delete customer record
                delete_query = "DELETE FROM CUSTOMER WHERE IdNo = %s"
                cursor.execute(delete_query, (customer_id,))
                connection.commit()
                messagebox.showinfo("Success", "Customer record deleted successfully")
                clear_entries()
        except Error as e:
            messagebox.showerror("Error", f"Error while deleting customer record: {e}")

    def retrieve_customer():
        customer_id = entry_customer_id.get()
        
        if not customer_id:
            messagebox.showerror("Error", "Customer ID must be provided")
            return
        
        try:
            connection = connect_to_database()
            if connection:
                cursor = connection.cursor()
                # Retrieve customer record
                cursor.execute("SELECT * FROM CUSTOMER WHERE IdNo = %s", (customer_id,))
                customer_data = cursor.fetchone()
                if not customer_data:
                    messagebox.showerror("Error", "Customer with the provided ID does not exist")
                    return
                clear_entries()
                # Display customer data in a table-like format
                display_customer_data(customer_data)
        except Error as e:
            messagebox.showerror("Error", f"Error while retrieving customer record: {e}")

    def clear_entries():
        entry_customer_id.delete(0, tk.END)
        entry_first_name.delete(0, tk.END)
        entry_last_name.delete(0, tk.END)
        entry_address.delete(0, tk.END)

    def display_customer_data(customer_data):
        # Create a new window to display customer data
        customer_window = tk.Toplevel()
        customer_window.title("Customer Data")
        customer_window.configure(bg='#c6e2e9')  # Set the pastel background color

        # Create a table-like display using ttk.Treeview
        customer_table = ttk.Treeview(customer_window, columns=("ID", "First Name", "Last Name", "Address"))
        customer_table.heading("#1", text="ID", anchor=tk.W)
        customer_table.heading("#2", text="First Name", anchor=tk.W)
        customer_table.heading("#3", text="Last Name", anchor=tk.W)
        customer_table.heading("#4", text="Address", anchor=tk.W)
        customer_table.pack(padx=10, pady=10)

        # Insert customer data into the table
        customer_table.insert("", "end", values=customer_data)

    def retrieve_all_customers():
        try:
            connection = connect_to_database()
            if connection:
                cursor = connection.cursor()
                # Retrieve all customer records
                cursor.execute("SELECT * FROM CUSTOMER")
                customer_data = cursor.fetchall()
                if not customer_data:
                    messagebox.showinfo("Info", "No customers found in the database")
                    return
                
                # Create a new window to display all customers
                all_customers_window = tk.Toplevel()
                all_customers_window.title("All Customers")
                all_customers_window.configure(bg='#c6e2e9')  # Set the pastel background color

                # Create a table-like display using ttk.Treeview
                customer_table = ttk.Treeview(all_customers_window, columns=("ID", "First Name", "Last Name", "Address"))
                customer_table.heading("#1", text="ID", anchor=tk.W)
                customer_table.heading("#2", text="First Name", anchor=tk.W)
                customer_table.heading("#3", text="Last Name", anchor=tk.W)
                customer_table.heading("#4", text="Address", anchor=tk.W)
                customer_table.pack(padx=10, pady=10)

                # Insert customer data into the table
                for customer in customer_data:
                    customer_table.insert("", "end", values=customer)
                
        except Error as e:
            messagebox.showerror("Error", f"Error while retrieving customer records: {e}")

    root = tk.Tk()
    root.title("Customer Management")
    root.configure(bg='#c6e2e9')  # Set the pastel background color

    # Labels
    label_customer_id = tk.Label(root, text="Customer ID:", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black")
    label_first_name = tk.Label(root, text="First Name:", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black")
    label_last_name = tk.Label(root, text="Last Name:", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black")
    label_address = tk.Label(root, text="Address:", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black")

    label_customer_id.grid(row=0, column=0, padx=10, pady=5, sticky="e")
    label_first_name.grid(row=1, column=0, padx=10, pady=5, sticky="e")
    label_last_name.grid(row=2, column=0, padx=10, pady=5, sticky="e")
    label_address.grid(row=3, column=0, padx=10, pady=5, sticky="e")

    # Entry fields
    global entry_customer_id
    global entry_first_name
    global entry_last_name
    global entry_address

    entry_customer_id = tk.Entry(root, width=30)  # Increase the width of the text boxes
    entry_first_name = tk.Entry(root, width=30)    # Increase the width of the text boxes
    entry_last_name = tk.Entry(root, width=30)    # Increase the width of the text boxes
    entry_address = tk.Entry(root, width=30)    # Increase the width of the text boxes

    entry_customer_id.grid(row=0, column=1, padx=10, pady=5)
    entry_first_name.grid(row=1, column=1, padx=10, pady=5)
    entry_last_name.grid(row=2, column=1, padx=10, pady=5)
    entry_address.grid(row=3, column=1, padx=10, pady=5)

    # Buttons
    insert_button = tk.Button(root, text="Insert", command=insert_customer, font=("Arial", 12), bg='#c6e2e9', fg="black")
    update_button = tk.Button(root, text="Update", command=update_customer, font=("Arial", 12), bg='#c6e2e9', fg="black")
    delete_button = tk.Button(root, text="Delete", command=delete_customer, font=("Arial", 12), bg='#c6e2e9', fg="black")
    retrieve_button = tk.Button(root, text="Retrieve", command=retrieve_customer, font=("Arial", 12), bg='#c6e2e9', fg="black")
    retrieve_all_button = tk.Button(root, text="Retrieve_all", command=retrieve_all_customers, font=("Arial", 12), bg='#c6e2e9', fg="black")
    clear_button = tk.Button(root, text="Clear", command=clear_entries, font=("Arial", 12), bg='#c6e2e9', fg="black")

    insert_button.grid(row=4, column=0, padx=10, pady=10)
    update_button.grid(row=4, column=1, padx=10, pady=10)
    delete_button.grid(row=5, column=0, padx=10, pady=10)
    retrieve_button.grid(row=5, column=1, padx=10, pady=10)
    retrieve_all_button.grid(row=6, column=0, padx=10, pady=10)
    clear_button.grid(row=6, column=1, columnspan=2, padx=10, pady=10)

    root.mainloop()

if __name__ == "__main__":
    customer()
```

# DAILY NEWS SUBSCRIPTION

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import mysql.connector
from mysql.connector import Error
from datetime import datetime, timedelta

# Function to calculate end date
def calculate_end_date(start_date, no_of_days):
    try:
        start_date_obj = datetime.strptime(start_date, "%Y-%m-%d")
        no_of_days = int(no_of_days)
        end_date_obj = start_date_obj + timedelta(days=no_of_days)
        return end_date_obj.strftime("%Y-%m-%d")
    except ValueError:
        return None

# Function to fetch the daily rate based on newspaper name and subscription duration
def calculate_price(d_name, no_of_days):
    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor(buffered=True)
            # Execute the SQL query to fetch the daily rate for the specified newspaper and number of days
            cursor.execute("SELECT Daily_Rate FROM DAILY_NEWS_PRICING WHERE D_Name = %s AND No_Of_Days = %s", (d_name, no_of_days))
            daily_rate = cursor.fetchone()
            # Return the daily rate if found
            if daily_rate:
                return float(daily_rate[0])
    except Error as e:
        messagebox.showerror("Error", f"Error while fetching daily rate: {e}")
    finally:
        if connection:
            connection.close()
    return None


# Insert operation function
def insert_daily_news_subscription(id_no, start_date, d_name, no_of_days):
    # Calculate End Date and Price
    end_date = calculate_end_date(start_date, no_of_days)
    price = calculate_price(d_name, no_of_days)

    if end_date is None or price is None:
        messagebox.showerror("Error", "Failed to calculate End Date or Price")
        return

    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor()
            insert_query = "INSERT INTO DAILY_NEWS_SUBSCRIPTION (IdNo, Start_Date, End_Date, D_Name, No_Of_Days, Price) VALUES (%s, %s, %s, %s, %s, %s)"
            cursor.execute(insert_query, (id_no, start_date, end_date, d_name, no_of_days, price))
            connection.commit()
            messagebox.showinfo("Success", "Daily News Subscription record inserted successfully")
    except Error as e:
        messagebox.showerror("Error", f"Error while inserting Daily News Subscription record: {e}")
    finally:
        if connection:
            connection.close()

# Delete operation function
def delete_daily_news_subscription(id_no, d_name):
    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor()
            delete_query = "DELETE FROM DAILY_NEWS_SUBSCRIPTION WHERE IdNo = %s AND D_Name = %s"
            cursor.execute(delete_query, (id_no, d_name))
            connection.commit()
            if cursor.rowcount > 0:
                messagebox.showinfo("Success", "Daily News Subscription record deleted successfully")
            else:
                messagebox.showerror("Error", "Daily News Subscription record not found")
    except Error as e:
        messagebox.showerror("Error", f"Error while deleting Daily News Subscription record: {e}")
    finally:
        if connection:
            connection.close()

# Retrieve operation function for a single daily news subscription
def retrieve_daily_news_subscription(id_no, d_name):
    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM DAILY_NEWS_SUBSCRIPTION WHERE IdNo = %s AND D_Name = %s", (id_no, d_name))
            data = cursor.fetchone()
            if data:
                # Create a new window to display the data
                single_data_window = tk.Toplevel()
                single_data_window.title("Daily News Subscription Data")

                # Create a Treeview widget
                table = ttk.Treeview(single_data_window, columns=("IdNo", "Start_Date", "End_Date", "D_Name", "No_Of_Days", "Price"), show='headings')
                for col in table['columns']:
                    table.heading(col, text=col)

                # Insert data into the table
                table.insert("", "end", values=data)

                table.pack(expand=True, fill='both')
            else:
                messagebox.showinfo("Info", "Daily News Subscription record not found")
    except Error as e:
        messagebox.showerror("Error", f"Error while retrieving Daily News Subscription record: {e}")
    finally:
        if connection:
            connection.close()

# Retrieve all operation function
def retrieve_all_daily_news_subscription():
    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor()
            cursor.execute("SELECT * FROM DAILY_NEWS_SUBSCRIPTION")
            data = cursor.fetchall()
            if data:
                # Show the data in a new window or a table
                all_data_window = tk.Toplevel()
                all_data_window.title("All Daily News Subscriptions")
                table = ttk.Treeview(all_data_window, columns=("IdNo", "Start_Date", "End_Date", "D_Name", "No_Of_Days", "Price"), show='headings')
                for col in table['columns']:
                    table.heading(col, text=col)
                for row in data:
                    table.insert("", "end", values=row)
                table.pack(expand=True, fill='both')
            else:
                messagebox.showinfo("Info", "No Daily News Subscription records found")
    except Error as e:
        messagebox.showerror("Error", f"Error while retrieving all Daily News Subscription records: {e}")
    finally:
        if connection:
            connection.close()

# Additional import for partial function
from functools import partial

# Update operation function
def update_daily_news_subscription(id_no, start_date, d_name, no_of_days):
    end_date = calculate_end_date(start_date, no_of_days)
    price = calculate_price(d_name, no_of_days)

    if end_date is None or price is None:
        messagebox.showerror("Error", "Failed to calculate End Date or Price")
        return

    try:
        connection = connect_to_database()
        if connection:
            cursor = connection.cursor()
            update_query = "UPDATE DAILY_NEWS_SUBSCRIPTION SET Start_Date = %s, End_Date = %s, No_Of_Days = %s, Price = %s WHERE IdNo = %s AND D_Name = %s"
            cursor.execute(update_query, (start_date, end_date, no_of_days, price, id_no, d_name))
            connection.commit()
            if cursor.rowcount > 0:
                messagebox.showinfo("Success", "Daily News Subscription record updated successfully")
            else:
                messagebox.showerror("Error", "Daily News Subscription record not found")
    except Error as e:
        messagebox.showerror("Error", f"Error while updating Daily News Subscription record: {e}")
    finally:
        if connection:
            connection.close()

# Clear entries function
def clear_entries(entries):
    for entry in entries:
        entry.delete(0, tk.END)

# GUI components with updated visual style
def daily_news_subscription():
    root = tk.Tk()
    root.title("Daily News Subscription Management")
    root.configure(bg='#c6e2e9')  # Set the pastel background color

    # Entry widgets
    entry_id_no = tk.Entry(root, width=30)
    entry_start_date = tk.Entry(root, width=30)
    entry_d_name = tk.Entry(root, width=30)
    entry_no_of_days = tk.Entry(root, width=30)

    # Collect all entries for easy clearing
    entries = [entry_id_no, entry_start_date, entry_d_name, entry_no_of_days]

    # Labels with specified font and color
    tk.Label(root, text="Customer ID", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black").grid(row=0, column=0, padx=10, pady=5, sticky="e")
    tk.Label(root, text="Start Date (YYYY-MM-DD)", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black").grid(row=1, column=0, padx=10, pady=5, sticky="e")
    tk.Label(root, text="Newspaper Name", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black").grid(row=2, column=0, padx=10, pady=5, sticky="e")
    tk.Label(root, text="Number of Days", font=("Arial", 12, "bold"), bg='#c6e2e9', fg="black").grid(row=3, column=0, padx=10, pady=5, sticky="e")

    # Place Entry widgets
    entry_id_no.grid(row=0, column=1, padx=10, pady=5)
    entry_start_date.grid(row=1, column=1, padx=10, pady=5)
    entry_d_name.grid(row=2, column=1, padx=10, pady=5)
    entry_no_of_days.grid(row=3, column=1, padx=10, pady=5)

    # Buttons with specified font and color
    tk.Button(root, text="Insert", command=lambda: insert_daily_news_subscription(entry_id_no.get(), entry_start_date.get(), entry_d_name.get(), entry_no_of_days.get()), font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=4, column=0, padx=10, pady=10)
    tk.Button(root, text="Update", command=lambda: update_daily_news_subscription(entry_id_no.get(), entry_start_date.get(), entry_d_name.get(), entry_no_of_days.get()), font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=4, column=1, padx=10, pady=10)
    tk.Button(root, text="Delete", command=lambda: delete_daily_news_subscription(entry_id_no.get(), entry_d_name.get()), font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=5, column=0, padx=10, pady=10)
    tk.Button(root, text="Retrieve", command=lambda: retrieve_daily_news_subscription(entry_id_no.get(), entry_d_name.get()), font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=5, column=1, padx=10, pady=10)
    tk.Button(root, text="Retrieve All", command=retrieve_all_daily_news_subscription, font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=6, column=0, padx=10, pady=10)
    tk.Button(root, text="Clear", command=partial(clear_entries, entries), font=("Arial", 12), bg='#c6e2e9', fg="black").grid(row=6, column=1, padx=10, pady=10)

    root.mainloop()

if __name__ == "__main__":
    daily_news_subscription()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk

# Function to open a table window for a selected table
def op_table(table_name):
    if table_name == "customer":
        customer()
    elif table_name == "publication":
        publication()
    elif table_name == "newspaper":
        newspaper()
    elif table_name == "magazine":
        magazine()
    elif table_name == "magazine_pricing":
        magazine_pricing()
    elif table_name == "daily_news_pricing":
        daily_news_pricing()
    elif table_name == "weekly_news_pricing":
        weekly_news_pricing()
    elif table_name == "daily_news_subscription":
        daily_news_subscription()  # Call the daily_news_subscription function
    elif table_name == "weekly_news_subscription":
        weekly_news_subscription()  # Call the weekly_news_subscription function
    elif table_name == "magazine_subscription":
        magazine_subscription()  # Call the magazine_subscription function
    else:
        table_window = tk.Toplevel()
        table_window.title(f"{table_name} Table")
        table_window.configure(bg='#c6e2e9')
        label_table = tk.Label(table_window, text=f"{table_name} Table Page", font=("Arial", 14, "bold"), fg="black", bg='#c6e2e9')
        label_table.pack(pady=20)


def main():
    root = tk.Tk()
    root.title("Database Management")
    root.geometry("800x600")  # Set the initial window size (you can adjust this)
    root.configure(bg='#c6e2e9')  # Set the pastel background color

    # Create a title label
    title_label = tk.Label(root, text="Subscription Management System", font=("Arial", 16, "bold"), fg="black", bg='#c6e2e9')
    title_label.pack(pady=20)

    # Create buttons for each table
    tables = ["customer", "publication", "newspaper", "magazine", "daily_news_pricing","magazine_pricing",
               "weekly_news_pricing", "daily_news_subscription", "weekly_news_subscription",
                "magazine_subscription"]
            

    for table in tables:
        table_button = tk.Button(root, text=table.capitalize(), command=lambda t=table: op_table(t), font=("Arial", 12), bg='#c6e2e9', fg="black")
        table_button.pack(padx=10, pady=5, fill=tk.X)

    root.mainloop()

if __name__ == "__main__":
    main()
```
